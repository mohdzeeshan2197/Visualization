# Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. rel plot    # relational plot
2. scatter plot
3. dis plot 
4. pair plot
5. joint plot

```python
import seaborn as sns 
```

```python
#                  ****List of Data-set Insid seaborn*****

sns.get_dataset_names()
```

```python
# 1. rel plot
#                           ***for example we take a dataset from seaborn*** 

taxi = sns.load_dataset('taxis')
```

```python
taxi
```

```python
sns.relplot(x=taxi['distance'], y= taxi['fare'])
```

```python
sns.relplot(x=taxi['fare'], y=taxi['tip'], style=taxi['passengers'])
```

```python
sns.catplot(x= 'passengers', y= 'color', data= taxi)
```

```python
# 2                ****scatter Plot****


sns.scatterplot(x=taxi['fare'],y = taxi['tip'])
```

```python
# 3. dis plot


sns.displot(taxi['payment'])
```

```python
#4. pair plot

sns.pairplot(taxi)
```

```python
#  5 joint plot

sns.jointplot(x=taxi['fare'],y = taxi['tip'])
```
